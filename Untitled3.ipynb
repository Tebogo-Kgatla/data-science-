```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import missingno as msno
from plotly.subplots import make_subplots
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 6)
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# [CELL 3] - Load and explore the dataset
from google.colab import files
import io

# Upload the dataset
uploaded = files.upload('diabetic_data.csv')

# Load the data
filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[filename]))

print("üîç DATASET OVERVIEW")
print("="*50)
print(f"Dataset Shape: {df.shape}")
print(f"Number of Patients: {len(df)}")
print(f"Number of Features: {len(df.columns)}")

# Display basic info
print("\nüìä DATA TYPES:")
print(df.dtypes.value_counts())

print("\nüëÄ FIRST FEW ROWS:")
display(df.head())
```

```python
# [CELL 4] - Data Quality Dashboard
print("üéØ DATA QUALITY ASSESSMENT")
print("="*50)

# Create a comprehensive data quality visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Missing values matrix
msno.matrix(df, ax=axes[0,0])
axes[0,0].set_title('Missing Values Pattern', fontsize=14, fontweight='bold')

# 2. Data types distribution
df.dtypes.value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[0,1])
axes[0,1].set_title('Feature Types Distribution', fontsize=14, fontweight='bold')

# 3. Missing values count
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0]
if len(missing_data) > 0:
    missing_data.plot(kind='bar', ax=axes[1,0])
    axes[1,0].set_title('Missing Values by Feature', fontsize=14, fontweight='bold')
else:
    axes[1,0].text(0.5, 0.5, 'No Missing Values!', ha='center', va='center', fontsize=16)
    axes[1,0].set_title('Missing Values by Feature', fontsize=14, fontweight='bold')

# 4. Dataset size info
info_data = {'Patients': len(df), 'Features': len(df.columns)}
pd.Series(info_data).plot(kind='bar', ax=axes[1,1], color=['skyblue', 'lightcoral'])
axes[1,1].set_title('Dataset Dimensions', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Print summary statistics
print(f"‚úÖ Total missing values: {df.isnull().sum().sum()}")
print(f"‚úÖ Duplicate records: {df.duplicated().sum()}")
```

```python
# [CELL 5] - Data Cleaning Pipeline Visualization
print("üßπ DATA CLEANING PROCESS")
print("="*50)

# Store cleaning steps for visualization
cleaning_steps = []
records_count = []

# Step 0: Original data
cleaning_steps.append("Original Data")
records_count.append(len(df))
original_missing = df.isnull().sum().sum()

# Create a copy for processing
df_clean = df.copy()

# Step 1: Handle missing values
print("1. Handling missing values...")
for col in df_clean.columns:
    if df_clean[col].isnull().sum() > 0:
        if df_clean[col].dtype == 'object':
            df_clean[col].fillna('Unknown', inplace=True)
        else:
            df_clean[col].fillna(df_clean[col].median(), inplace=True)

cleaning_steps.append("After Missing Value\nTreatment")
records_count.append(len(df_clean))

# Step 2: Convert target variable to binary
print("2. Processing target variable...")
df_clean['readmitted_binary'] = df_clean['readmitted'].apply(
    lambda x: 1 if x == '<30' else 0
)
cleaning_steps.append("Target Variable\nEncoded")
records_count.append(len(df_clean))

# Step 3: Encode categorical variables
print("3. Encoding categorical variables...")
categorical_cols = df_clean.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_cols:
    if col != 'readmitted':  # We already handled the target
        le = LabelEncoder()
        df_clean[col] = le.fit_transform(df_clean[col].astype(str))
        label_encoders[col] = le

cleaning_steps.append("Categorical Features\nEncoded")
records_count.append(len(df_clean))

# Visualize the cleaning pipeline
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Pipeline progress
ax1.bar(cleaning_steps, records_count, color=['red', 'orange', 'yellow', 'green'])
ax1.set_title('Data Cleaning Pipeline Progress', fontsize=14, fontweight='bold')
ax1.set_ylabel('Number of Records')
ax1.tick_params(axis='x', rotation=45)

# Missing values reduction
missing_progress = [original_missing, df_clean.isnull().sum().sum(), 0, 0]
ax2.plot(cleaning_steps, missing_progress, marker='o', linewidth=3, markersize=8)
ax2.set_title('Missing Values Reduction', fontsize=14, fontweight='bold')
ax2.set_ylabel('Missing Values Count')
ax2.tick_params(axis='x', rotation=45)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"‚úÖ Final dataset shape: {df_clean.shape}")
print(f"‚úÖ Remaining missing values: {df_clean.isnull().sum().sum()}")
```

```python
# [CELL 6] - Interactive Exploratory Data Analysis
print("üìà INTERACTIVE EXPLORATORY DATA ANALYSIS")
print("="*50)

# Create interactive visualizations
fig = go.Figure()

# 1. Target variable distribution
target_counts = df['readmitted'].value_counts()
fig.add_trace(go.Bar(x=target_counts.index, y=target_counts.values,
                     name='Readmission Status', marker_color='indianred'))

fig.update_layout(title='Patient Readmission Distribution',
                  xaxis_title='Readmission Status',
                  yaxis_title='Count')
fig.show()

# 2. Interactive feature distributions
feature = 'time_in_hospital'  # You can make this interactive with widgets
fig = px.histogram(df, x=feature, color='readmitted',
                   title=f'Distribution of {feature} by Readmission Status',
                   barmode='overlay', opacity=0.7)
fig.show()

# 3. Correlation heatmap (for numerical features)
print("üîó FEATURE CORRELATION HEATMAP")
numerical_cols = df_clean.select_dtypes(include=[np.number]).columns
corr_matrix = df_clean[numerical_cols].corr()

fig = px.imshow(corr_matrix,
                title='Feature Correlation Matrix',
                color_continuous_scale='RdBu_r',
                aspect='auto')
fig.show()
```

```python
# [CELL 8] - Interactive Feature Explorer
from ipywidgets import interact, Dropdown

@interact
def explore_feature(feature=Dropdown(
    options=['time_in_hospital', 'num_medications', 'num_lab_procedures', 'age'],
    value='time_in_hospital',
    description='Feature:'
)):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Distribution by readmission status
    for status in df['readmitted'].unique():
        subset = df[df['readmitted'] == status]
        ax1.hist(subset[feature], alpha=0.7, label=status, bins=20)

    ax1.set_title(f'Distribution of {feature} by Readmission Status')
    ax1.legend()
    ax1.set_xlabel(feature)
    ax1.set_ylabel('Frequency')

    # Box plot
    df.boxplot(column=feature, by='readmitted', ax=ax2)
    ax2.set_title(f'{feature} vs Readmission Status')

    plt.suptitle('')
    plt.tight_layout()
    plt.show()
```
